```python
def line(x, m, b): 
    y = m * x + b
    return y

xpoints = np.arange(10)
ypoints = line(xpoints, 2, -3)

print(xpoints, ypoints)
```

```python
plt.plot(xpoints,ypoints)
```

```python
plt.plot(xpoints,ypoints)

plt.axis([-3, 10, -7, 18])

plt.title('distance vs velocity')
plt.xlabel('velocity')
plt.ylabel('distance')
```
